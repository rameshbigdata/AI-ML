```python
#in pandas ----> Apply

def upper_case_title(title):
  return  title.upper()

df['review_title_uppercase']= df['review_title'].apply(upper_case_title)
```

```python
df['review_title_uppercase']= df['review_title'].apply(lambda x: x.upper())
```

```python
def extract_price(price_str):
    try:
        return float(str(price_str).replace('₹', '').replace(',', ''))
    except (ValueError, TypeError):
        return None
```

```python
df['discounted_price_fixed'] = df['discounted_price'].apply(extract_price)
df['actual_price_fixed'] = df['actual_price'].apply(extract_price)
```

```python
df.head()
```

```python
#feature engineering

def calculate_discount_percentage(xy):
    return ((xy['actual_price_fixed'] - xy['discounted_price_fixed']) / xy['actual_price_fixed']) * 100


df['calculated_discount'] = df.apply(calculate_discount_percentage, axis=1)
```

```python
df['calculated_discount_alternative'] = ((df['actual_price_fixed'] - df['discounted_price_fixed']) / df['actual_price_fixed']) * 100
```

```python
#APPLY ON GROUPED DATA?

def avg_price_diff(group):
    avg_actual = group['actual_price_fixed'].mean()
    avg_discounted = group['discounted_price_fixed'].mean()

    return avg_actual - avg_discounted

df.groupby('category').apply(avg_price_diff)
```

```python
#can we apply for filtering
def add_review_status(group):
    group['review_status'] = group['rating'].apply(lambda x: 'Good' if x is not None and x>4.2 else 'Not Good')
    return group

df.groupby('category').apply(add_review_status)
```
