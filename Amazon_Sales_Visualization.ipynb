```python
import pandas as pd
import numpy as np


df = pd.read_csv('clean_amazon_data.csv')
df.head()
```

```python
df['rating'].mean()
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
sns.histplot(x='rating', data= df)
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.title('Distribution of Ratings')

plt.axvline(x= df['rating'].mean(), color='r', linestyle='--', label=f'Mean Rating: {df["rating"].mean():.2f}')
plt.legend()

plt.show() #remove the metadata and just show the chart
```

```python

plt.figure(figsize=(8,6))
sns.barplot(x='category', y='rating', data=df, errorbar=None)
plt.xticks(rotation=90)
plt.show()
```

```python

sns.kdeplot(
    data=df,
    x="discounted_price",
    y="actual_price"
)
```

```python
#Bi-variate Analysis

#2 numerical variables

#Scatterplot


plt.figure(figsize=(8,6) )
sns.scatterplot(x="discounted_price", y="actual_price", hue='category' , data=df)
```
